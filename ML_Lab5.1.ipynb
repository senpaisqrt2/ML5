## 2. При горизонтальной интеграции первой и второй таблицы выведите по каждому столбцу гистограмму распределения или плотность распределения, чтобы убедиться, что признаки выражаются по соотносимым шкалам.

```python
import matplotlib.pyplot as plt
```

```python
numeric_cols = data1.select_dtypes(include=['int64', 'float64']).columns.union(
               data2.select_dtypes(include=['int64', 'float64']).columns)
```

```python
for col in numeric_cols:
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    sns.histplot(data1[col], color='blue', alpha=0.5, label='Data1', kde=True)
    sns.histplot(data2[col], color='orange', alpha=0.5, label='Data2', kde=True)
    plt.title(f'Histogram of {col}')
    plt.legend()

    plt.subplot(1, 2, 2)
    sns.kdeplot(data1[col], color='blue', label='Data1')
    sns.kdeplot(data2[col], color='orange', label='Data2')
    plt.title(f'Density Plot of {col}')
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
categorical_cols = data1.select_dtypes(include=['object']).columns.union(
                  data2.select_dtypes(include=['object']).columns).drop('Customer Remarks')
```

```python
for col in categorical_cols:
    plt.figure(figsize=(12, 6))

    counts1 = data1[col].value_counts(normalize=True)
    counts2 = data2[col].value_counts(normalize=True)

    plot_df = pd.DataFrame({'Data1': counts1, 'Data2': counts2}).fillna(0)

    plot_df.plot(kind='bar', ax=plt.gca())
    plt.title(f'Distribution of {col}')
    plt.ylabel('Proportion')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

## 3. При выполнении третьего задания убедитесь более явно, что обильное количество пропущенных значений не является ошибкой объединения. Проведите больше, чем одну точечную проверку.

```python
import missingno as msno
```

```python
def check_na(df, df_name):
    na_percent = df.isna().mean() * 100
    print(f"\nПропуски в {df_name}:")
    print(na_percent[na_percent > 0].sort_values(ascending=False))
```

```python
check_na(data_12, "data_12 (data1 + data2)")
check_na(data3, "data3")
```

```python
check_na(data_123, "Объединённые данные (data_123)")
```

## 4. При выполнении третьего задания попробуйте использовать разные виды соединений. Как это отражается на структуре датасета?

```python
from matplotlib_venn import venn2
```

```python
print(f"Уникальные ID в data_12: {data_12['Id'].nunique()}")
print(f"Уникальные ID в data3: {data3['Id'].nunique()}")
print(f"Пересекающиеся ID: {len(set(data_12['Id']).intersection(set(data3['Id'])))}")
```

```python
inner_merge = data_12.merge(data3, on='Id', how='inner')
```

```python
outer_merge = data_12.merge(data3, on='Id', how='outer', indicator=True)
```

```python
left_merge = data_12.merge(data3, on='Id', how='left')
```

```python
right_merge = data_12.merge(data3, on='Id', how='right')
```

```python
results = pd.DataFrame({
    'Тип соединения': ['inner', 'outer', 'left', 'right'],
    'Строк': [
        len(inner_merge),
        len(outer_merge),
        len(left_merge),
        len(right_merge)
    ],
    'Уникальных ID': [
        inner_merge['Id'].nunique(),
        outer_merge['Id'].nunique(),
        left_merge['Id'].nunique(),
        right_merge['Id'].nunique()
    ],
    'Пропуски в Agent_name': [
        inner_merge['Agent_name_x'].isna().mean(),
        outer_merge['Agent_name_x'].isna().mean(),
        left_merge['Agent_name_x'].isna().mean(),
        right_merge['Agent_name_x'].isna().mean()
    ]
})
print(results)
```

```python
venn2([set(data_12['Id']), set(data3['Id'])],
      set_labels=('data_12', 'data3'))
plt.title("Пересечение ID между таблицами")
plt.show()
```
