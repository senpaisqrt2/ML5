```python
categorical_cols = data1.select_dtypes(include=['object']).columns.union(
                  data2.select_dtypes(include=['object']).columns).drop('Customer Remarks')
```

```python
for col in categorical_cols:
    plt.figure(figsize=(12, 6))

    counts1 = data1[col].value_counts(normalize=True)
    counts2 = data2[col].value_counts(normalize=True)

    plot_df = pd.DataFrame({'Data1': counts1, 'Data2': counts2}).fillna(0)

    plot_df.plot(kind='bar', ax=plt.gca())
    plt.title(f'Distribution of {col}')
    plt.ylabel('Proportion')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
